```python
import requests
# 
url = "http://127.0.0.1:8000/hybrid_search"
params = {"query": "RAG", "alpha": 0.5}

try:
    response = requests.get(url, params=params)
    if response.status_code == 200:
        print("✅ connection success！API return rsult:")
        print(response.json())
    else:
        print(f"❌ error: API  {response.status_code}")
except Exception as e:
    print(f"❌ can't connect API: {e}")
```

```python
import pandas as pd

# 1. 
queries = [
    "What is hybrid retrieval?",
    "What are the advantages of the model proposed in the paper?",
    "How to optimize the FAISS index?"
]
# 2. 
def run_evaluation(queries):
    eval_results = []
    
    for q in queries:
        # adjust alpha from 0.1-0.9
        response = requests.get(f"http://127.0.0.1:8000/hybrid_search", params={"query": q, "alpha": 0.5})
        
        if response.status_code == 200:
            data = response.json()
            # analysis
            eval_results.append({
                "query": q,
                "result": str(data.get("results", "No results")),
                "status": "Success"
            })
        else:
            eval_results.append({"query": q, "result": None, "status": "Failed"})
            
    return pd.DataFrame(eval_results)

# 3. 
df_results = run_evaluation(queries)
print("result---:")
print(df_results)
```

```python
import requests
import pandas as pd

# query key setting


# 設定關鍵字
ground_truth = [
    {
        "query": "How are pose features defined for each superpoint?", 
        "keywords": ["normal vector", "principal component", "cosine"]
    },
    {
        "query": "What are the limitations of superpoint-based methods?", 
        "keywords": ["preprocessing", "costly", "expressivity"]
    }
]

def calculate_precision_at_k(results, keywords):
    # 
    hits = sum(1 for kw in keywords if kw in str(results))
    return hits / len(keywords)

results_log = []

for item in ground_truth:
    response = requests.get("http://127.0.0.1:8000/hybrid_search", params={"query": item['query']})
    if response.status_code == 200:
        data = response.json()
        score = calculate_precision_at_k(data['results'], item['keywords'])
        results_log.append({"query": item['query'], "score": score, "raw": data['results']})

# 顯示評測報告
df = pd.DataFrame(results_log)
print("score report:")
display(df[['query', 'score']])
```

```python
# 
ground_truth = [
    {"query": "What are the core principles of RAG?", "keywords": ["Retrieval", "Generation"]},
    {"query": "How to optimize the FAISS index?", "keywords": ["FAISS", "index"]},
]

def calculate_precision_at_k(results, keywords):
    # compaer content
    content = str(results).lower()
    hits = sum(1 for kw in keywords if kw.lower() in content)
    return hits / len(keywords)

results_log = []

for item in ground_truth:
    response = requests.get("http://127.0.0.1:8000/hybrid_search", params={"query": item['query']})
    if response.status_code == 200:
        data = response.json()
        score = calculate_precision_at_k(data['results'], item['keywords'])
        results_log.append({"query": item['query'], "score": score, "raw": data['results']})

# 顯示評測報告
import pandas as pd
df = pd.DataFrame(results_log)
print("評測分數報告:")
display(df[['query', 'score']])

print(results_log[0]['raw'])
```
